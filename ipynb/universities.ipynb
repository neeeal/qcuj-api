```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_colleges_and_universities_in_the_Philippines"
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 
print(soup.prettify()) 
```

```python
soup.find('td').text.strip()
```

```python
soup.findAll('td')
```

```python
tables = soup.findAll('table',{'class':'wikitable'})
print(tables)
```

```python
universities =[]
```

```python
for table in tables:
    rows = soup.findAll('tr')
    for row in rows:
        university = row.find('td')  
        if university: 
            university_name = university.text.strip()
            if university_name not in universities:
                universities.append(university_name)
                print(university_name)
```

```python
with open('universities.json', 'w') as file:
    json.dump(universities, file)
```

```python
with open('universities.json', 'r') as file:
    universities_data = json.load(file)
```

```python
def search_universities(title):
    universities = [uni for uni in universities_data if title.lower() in uni.lower()]
    
    results = universities
    print(universities)
```
